# NEON Tree Crown RGB Visualization

Simple notebook for visualizing RGB crown data.

```python
# Import libraries and functions
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

import sys
sys.path.append('..')

from neon_tree_classification.core.visualization import plot_rgb
from neon_tree_classification.core.visualization import plot_hsi
```

```python
# Load training data
df = pd.read_csv('../training_data_clean.csv')
print(f"Loaded {len(df)} crown samples")
df.head()
```

```python
from neon_tree_classification.core.visualization import plot_rgb
from neon_tree_classification.core.visualization import plot_hsi
```

```python
# Plot first crown
sample_crown = df.iloc[0]
rgb_path = sample_crown['rgb_path']
print(f"Plotting: {rgb_path}")
plot_rgb(rgb_path)
```

```python
# HSI Visualization - Pseudo RGB
from pathlib import Path
import matplotlib.pyplot as plt
from neon_tree_classification.core.visualization import plot_hsi

if 'hsi_path' in df.columns and pd.notna(sample_crown['hsi_path']):
    hsi_path = sample_crown['hsi_path']
    if Path(hsi_path).exists():
        fig = plot_hsi(hsi_path)
        plt.show()
    else:
        print("HSI file not found")
else:
    print("HSI data not available")
```

```python
# HSI Visualization - Spectral Signatures
from neon_tree_classification.core.visualization import plot_hsi_spectra

if 'hsi_path' in df.columns and pd.notna(sample_crown['hsi_path']):
    hsi_path = sample_crown['hsi_path']
    if Path(hsi_path).exists():
        fig = plot_hsi_spectra(hsi_path, num_pixels=2)
        plt.show()
    else:
        print("HSI file not found")
else:
    print("HSI data not available")
```

```python
# LiDAR Visualization - CHM (Canopy Height Model)
from neon_tree_classification.core.visualization import plot_lidar

if 'lidar_path' in df.columns and pd.notna(sample_crown['lidar_path']):
    lidar_path = sample_crown['lidar_path']
    if Path(lidar_path).exists():
        fig = plot_lidar(lidar_path)
        plt.show()
    else:
        print("LiDAR file not found")
else:
    print("LiDAR data not available")
```
